```python
# Import necessary libraries
import pandas as pd
import numpy as np
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder

# ds1 = pd.read_csv('dwell_times.csv')
# run_data = pd.read_csv('run_times.csv')
# trips = pd.read_csv('trips.csv')
# dwell_sorted = pd.read_csv('dwell_sorted.csv')

#ds1=pd.concat((pd.read_csv("ds1"),pd.read_csv("ds3")))
ds1=pd.read_csv("ds1")

ds1 = ds1.sort_values(by=['trip_id', 'arrival_time'])

ds1['arrival_time'] = pd.to_datetime(ds1['arrival_time'])
ds1['previous_arrival_time'] = ds1.groupby('trip_id')['arrival_time'].shift(1)

ds1['time_diff_seconds'] = (ds1['arrival_time'] - ds1['previous_arrival_time']).dt.total_seconds()

ds1 = ds1.dropna(subset=['previous_arrival_time'])
```

```python
features = ['deviceid', 'direction', 'bus_stop', 'dwell_time_in_seconds']
target = 'time_diff_seconds'
```

```python
X = ds1[features]
y = ds1[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

```python
categorical_features = ['deviceid', 'direction', 'bus_stop']

le_deviceid = LabelEncoder()
le_bus_stop = LabelEncoder()
```

```python
X_train['deviceid_enc'] = le_deviceid.fit_transform(X_train['deviceid'])
X_train['bus_stop_enc'] = le_bus_stop.fit_transform(X_train['bus_stop'])
X_test['deviceid_enc'] = le_deviceid.transform(X_test['deviceid'])
X_test['bus_stop_enc'] = le_bus_stop.transform(X_test['bus_stop'])

X_train_xgb = X_train[['deviceid_enc', 'direction', 'bus_stop_enc', 'dwell_time_in_seconds']]
X_test_xgb = X_test[['deviceid_enc', 'direction', 'bus_stop_enc', 'dwell_time_in_seconds']]
```

```python
X_train_lgb = X_train[features]
X_test_lgb = X_test[features]

print("Training XGBoost model...")
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=100,
    max_depth=6,
    learning_rate=0.05,
    random_state=42,
    tree_method = "hist",
    device = "cuda"
)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
import numpy as np

# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [3, 5, 6, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3],
    'min_child_weight': [1, 5, 10]
}

# Convert data to DMatrix and specify GPU for training
dtrain = xgb.DMatrix(X_train_xgb, label=y_train)

# Define the XGBoost model, ensuring device is set to 'cuda'
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    tree_method='hist',
    random_state=42,
    device='cuda'
)

# Set up RandomizedSearchCV with cross-validation
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=50,
    scoring='neg_mean_absolute_error',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1,
)

# Fit the model using RandomizedSearchCV
random_search.fit(X_train_xgb, y_train)

# Print the best parameters and best score
print("Best parameters found: ", random_search.best_params_)
print("Best MAE score: ", -random_search.best_score_)
```

```python
random_search.fit(X_train_xgb, y_train)

# Print the best parameters and best score from the training search
print("Best parameters found: ", random_search.best_params_)
print("Best MAE score (cross-validation): ", -random_search.best_score_)

# Evaluate the best model on the test set
best_xgb_model = random_search.best_estimator_
y_pred_test = best_xgb_model.predict(X_test_xgb)

# Calculate evaluation metrics on the test set
mae_test = mean_absolute_error(y_test, y_pred_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

print("\nTest Set Evaluation:")
print(f"Mean Absolute Error: {mae_test:.2f} seconds")
print(f"Root Mean Squared Error: {rmse_test:.2f} seconds")
```

```python
print("\nTraining LightGBM model...")
lgb_train = lgb.Dataset(X_train_lgb, y_train, categorical_feature=categorical_features)
lgb_test = lgb.Dataset(X_test_lgb, y_test, reference=lgb_train, categorical_feature=categorical_features)

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'learning_rate': 0.1,
    'num_leaves': 31,
    'verbose': -1,
    'seed': 42
}

lgb_model = lgb.train(
    params,
    lgb_train,
    num_boost_round=100,
    valid_sets=[lgb_train, lgb_test],
    valid_names=['train', 'valid'],
    early_stopping_rounds=10,
    verbose_eval=10
)

y_pred_lgb = lgb_model.predict(X_test_lgb, num_iteration=lgb_model.best_iteration)
mae_lgb = mean_absolute_error(y_test, y_pred_lgb)
rmse_lgb = np.sqrt(mean_squared_error(y_test, y_pred_lgb))

print("\nLightGBM Model Evaluation:")
print(f"Mean Absolute Error: {mae_lgb:.2f} seconds")
print(f"Root Mean Squared Error: {rmse_lgb:.2f} seconds")
```
